```python
df=pd.read_csv("C:/Users/PC/02_Forward School/Capstone Project-Alcon Marshall/dataset _activity.csv")
```

```python
# Filter data from 2010 to 2017
df_filtered = df[(df['Year'] >= 2010) & (df['Year'] <= 2017)]

# Select only the 3 categories
categories = ['Food services', 'Event  catering services', 'Beverage services']
df_filtered = df_filtered[df_filtered['Activity'].isin(categories)]

# Pivot the data for easier plotting or tabular view
pivot_table = df_filtered.pivot(index='Year', columns='Activity', values='Gross Output')

# Display the pivot table
display(pivot_table)
```

```python
# Plotting
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
pivot_table.plot(marker='o')
plt.title("Value of Gross Output (20102017) by Category")
plt.ylabel("Gross Output (RM'000)")
plt.xlabel("Year")
plt.grid(True)
plt.tight_layout()
plt.legend(title='Activity')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Filter years 20102017
df_filtered = df[(df['Year'] >= 2010) & (df['Year'] <= 2017)]

# Step 2: Update industry list to include 5 categories
categories = [
    'Restaurants and restaurants cum night clubs',
    'Cafeterias/canteens',
    'Food stalls',
    'Fast food restaurants',
    'Ice cream trucks'  # Optional: replace or remove if needed
]
df_filtered = df_filtered[df_filtered['Industry'].isin(categories)]

# Step 3a: Pivot Gross Output
pivot_gross_output = df_filtered.pivot(index='Year', columns='Industry', values="Value of gross output (RM'000)")

# Step 3b: Pivot Number of Establishments
pivot_establishments = df_filtered.pivot(index='Year', columns='Industry', values="Number of establishments")

# Step 4: Plot Side-by-Side Comparison
fig, axes = plt.subplots(1, 2, figsize=(16, 6), sharex=True)

# Plot Gross Output
pivot_gross_output.plot(ax=axes[0], marker='o')
axes[0].set_title("Gross Output by Industry (RM'000)")
axes[0].set_ylabel("Gross Output")
axes[0].grid(True)

# Plot Establishments
pivot_establishments.plot(ax=axes[1], marker='s')
axes[1].set_title("Number of Establishments by Industry")
axes[1].set_ylabel("Establishments")
axes[1].grid(True)

# Common X label
for ax in axes:
    ax.set_xlabel("Year")

plt.suptitle("Gross Output vs Number of Establishments by Industry (20102017)", fontsize=14)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
